### spliting the data set (by slicing method) in order to explore the merge function

```python
data_set_01 = data_set.iloc[0:20000].reset_index(drop=True)
```

```python
data_set_02 = data_set.iloc[20000:].reset_index(drop=True)
```

```python
data_set_01
```

```python
data_set_02
```

```python
len(data_set) == len(data_set_01) + len(data_set_02)
```

```python
merge_inner = pd.merge(data_set_01, data_set_02, how='inner')
```

```python
len(merge_inner)
```

```python
merge_outer = pd.merge(data_set_01, data_set_02, how='outer')
```

```python
len(merge_outer)
```

#### inner merge can be likened to set intersect between the two data sets. i.e it only pics the data that is common to the two set. if not, you will get an emty set as above. while outer merge is like set union
